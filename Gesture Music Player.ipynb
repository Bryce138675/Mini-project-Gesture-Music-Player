```python
# Importing libraries
import cv2
import cvzone
from cvzone.HandTrackingModule import HandDetector  # Importing the hand tracking module
import numpy as np
import os
import time
import pygame
from pygame import mixer 
```

```python
#initialize the mixer module
pygame.mixer.init()
```

```python
# The following code reference from gesture Presentation PPT. (https://blog.csdn.net/dgvv4/article/details/123682731)
# Video capture
cap = cv2.VideoCapture(0)  # Get the computer camera
cap.set(3, 1280)  # Width of the image
cap.set(4, 720)  # Height of the image
```

```python
# Music dataset
folderpath = '/Users/liangminghao/Desktop/Gesture_music_player/music'
# Sorted by filename and stored in pathImage
pathImage = sorted(os.listdir(folderpath), key=len)  

imgNumeber = 0
```

```python
# Draw a line on the camera and make a gesture with your hand above the line to trigger
threshold = 350  # Height threshold for lines

buttonPressed = False   
buttonCounter = 0  
buttonDelay = 30  
```

```python
# Hand recognition module configuration
detector = HandDetector(maxHands=2, detectionCon=0.8)  # Up to 2 hands detected
```

```python
pTime=0
# The following code reference from gesture Presentation PPT. (https://blog.csdn.net/dgvv4/article/details/123682731)
while True:
 
    
    success, img = cap.read()
 
    # Flip the image so that the computer image is a mirror image of our own
    img = cv2.flip(img, flipCode=1)  
 
    # Return hand information - hands, image after drawing key point - img
    hands, img = detector.findHands(img)
 
    # Draw a line and make a gesture above the line to trigger
    cv2.line(img, (0,threshold), (1500,threshold), (0,255,0), 3)
    # Draw a finger movement mapping area
    cv2.rectangle(img, (400, 0), (1100,360), (255,0,0), 3)
 
    # Key point handling
    if hands and buttonPressed is False:  
        
        # Get all the information about a hand
        hand = hands[0]  
 
        # Counts how many fingers are cocked and returns a list where 1 means the finger is cocked and 0 means bent
        fingers = detector.fingersUp(hand)  
        print(fingers) 
 
 
        # Limit finger movement to a border
        lmList = hand['lmList']  # Get the xyz position of 21 key points
        # Setting the mapping area
        xval = int(np.interp(lmList[8][0], [300, 1280], [0, 1280]))  # x
        yval = int(np.interp(lmList[8][1], [0, 360], [0, 720]))  # y
 
        # Hand centre coordinates
        cx, cy = hand['center'] 
 
        # If the centre of the hand is above the line, the operation could continue
        if cy < threshold: 
            
            # The following code uses the pygame library to play music (https://blog.csdn.net/RNG_uzi_/article/details/98315639)
            
            # Use thumb and forefinger to play---Watermelon Sugar
            if fingers == [1,1,0,0,0]:
                mixer.init()
                mixer.music.load('/Users/liangminghao/Desktop/Gesture_music_player/music/Watermelon_Sugar.mp3')
                mixer.music.play()
                time.sleep(5)
                print('Watermelon Sugar')
                
            
            # Use thumbs, forefinger and little finger to play---I Knew You Were Trouble
            if fingers == [1,1,0,0,1]:
                mixer.init()
                mixer.music.load('/Users/liangminghao/Desktop/Gesture_music_player/music/I_Knew_You_Were_Trouble.mp3')
                mixer.music.play()
                time.sleep(5)
                print('I Knew You Were Trouble')
                
            
            # Use thumbs and little finger to play---Adore You
            if fingers == [1,0,0,0,1]:
                print('Adore You')
                mixer.init()
                mixer.music.load('/Users/liangminghao/Desktop/Gesture_music_player/music/Adore_You.mp3')
                mixer.music.play()
                time.sleep(5)
                
            
            # Use forefinger, middle finger, ring finger, little finger to play---As It Was
            if fingers == [0,1,1,1,1]:
                mixer.init()
                mixer.music.load('/Users/liangminghao/Desktop/Gesture_music_player/music/As_It_Was.mp3')
                mixer.music.play()
                time.sleep(5)
                print('As It Was')
                
            
            # Use forefinger and middle finger to play---Golden
            if fingers == [0,1,1,0,0]:
                mixer.init()
                mixer.music.load('/Users/liangminghao/Desktop/Gesture_music_player/music/Golden.mp3')
                mixer.music.play()
                time.sleep(5)
                print('Golden')
            
            # The following codes refer to the music player (https://blog.csdn.net/txr152111/article/details/119213827)
            # Use forefinger and little finger---pause music
            if fingers == [0,1,0,0,1]:
                pygame.mixer.music.pause()
                print('Pause')
            
            # Use thumb, middle finger, ring finger and little finger---unpause music
            if fingers == [1,0,1,1,1]:
                pygame.mixer.music.unpause()
                print('Unpause')
            
            # Use thumb, forefinger, middle finger, ring finger and little finger---stop music
            if fingers == [1,1,1,1,1]:
                pygame.mixer.music.stop()
                print('Stop')
            
            
    # Display
    cTime = time.time()  
    fps = 1 / (cTime - pTime)
    pTime = cTime
    print(fps)
    
    cv2.putText(img, str(int(fps)), (70, 50), cv2.FONT_HERSHEY_PLAIN, 3, (255, 0, 0), 3)

    cv2.imshow('frame', img)
    if cv2.waitKey(1) & 0xFF == 27:  # ESC button to exit
        break

# Release
cap.release()
cv2.destroyAllWindows()
```
